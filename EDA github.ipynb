```python
data = {
    'job_title': ['Engineer', 'Manager', 'Engineer', 'Manager'],
    'salary': ['₹ 4.5 - 7L/yr', '₹ 6 - 8L/yr', '₹ 3.5 - 6L/yr', '₹ 7 - 10L/yr']
}

df = pd.DataFrame(data)

# Function to clean salary data
def clean_salary(salary):
    # Remove currency symbols and 'yr' text
    salary = salary.replace('₹', '').replace('/yr', '').strip()
    
    # If it contains a range, take the average
    if ' - ' in salary:
        low, high = salary.split(' - ')
        low = float(low.strip().replace('L', '')) * 100000  # Convert to numeric
        high = float(high.strip().replace('L', '')) * 100000  # Convert to numeric
        return (low + high) / 2  # Return average of range
    else:
        return float(salary.replace('L', '').strip()) * 100000  # Convert to numeric

# Apply the function to the salary column
df['salary_numeric'] = df['salary'].apply(clean_salary)

# Now you can create the pivot table
pivot_table = df.pivot_table(index='job_title', values='salary_numeric', aggfunc='mean')
print(pivot_table)
```
